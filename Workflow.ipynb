This is a tutorial on how to use the RV Masses Code to simulate RV Data.

You will need a file for the target which includes information on the host star and a file for each individual planet. Examples on the formatting of these files are provided in the github.  

```python
import RVMassesCode
```

Here I am creating an object of the Target and passing in the Target example file.

In the next line I am loading in the planet data. 

```python
system = RVMassesCode.Target(file = 'target_ex_file.csv')
system.load_planets(planet_file = 'planet_ex_file.csv - Sheet1.csv')
```

Here I am creating an RV_obs object and passing in the system object from before into it.

```python
system_RV_obs = RVMassesCode.RV_obs(system)
```

Then I am inputting the dates where I will begin and end observing and the amount of times I will observe. I am also entering in a reference date. 

```python
system_RV_obs.sim_RVs(1, 50000, 15, 25000)
```

Next I can create the plots

```python
system_RV_obs.RV_plot()
```
